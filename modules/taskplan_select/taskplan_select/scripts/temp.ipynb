```python
import networkx as nx
import matplotlib.pyplot as plt


def get_object_color_from_type(type):
    return 'red'

def plot_graph(ax, nodes, edge_index):
    # Create a graph object
    G = nx.Graph()
    # Add nodes to the graph with labels
    node_colors = []
    for k, v in nodes.items():
        G.add_node(k, label=v['name'])
        node_colors.append(get_object_color_from_type(v['type']))
    # Add edges to the graph
    G.add_edges_from(edge_index)
    node_colors = ['red' for _ in nodes]
    node_labels = nx.get_node_attributes(G, 'label')
    # Draw the graph
    pos = nx.spring_layout(G)  # Positions for all nodes
    nx.draw(G, pos, ax, with_labels=True, labels=node_labels, node_color=node_colors, node_size=150,
            font_size=8, font_weight='regular', edge_color='black')
```

```python
graph = {'nodes': {0: {'name': 'A', 'type': [1, 0, 0 , 0]},
                   1: {'name': 'B', 'type': [0, 1, 0, 0]},
                   2: {'name': 'C', 'type': [0, 0, 1, 0]},
                   3: {'name': 'D', 'type': [0, 0, 0, 1]}},
         'edge_index': [(0, 1), (0, 2), (0, 3)]}
node_labels = [v['name'] for k, v in graph['nodes'].items()]
plt.figure(figsize=(10, 10))
ax = plt.subplot(111)
plot_graph(ax, graph['nodes'], graph['edge_index'])
```
